# Importing Libraries Set up Eviroment

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

## Loading Data into Pandas Dataframe

```python
Customers= pd.read_csv("Data/CSV/Customers.csv", encoding="latin1", sep=";")
Location= pd.read_csv("Data/CSV/Location.csv", encoding="latin1", sep=";")
Orders= pd.read_csv("Data/CSV/Orders.csv", encoding="latin1", sep=";")
Products= pd.read_csv("Data/CSV/Products.csv", encoding="latin1", sep=";")
```

### Pushing Dataframes Into SQL Sales Database

```python
conn = sqlite3.connect("Data/SQL_db/sales.db")
```

```python
Customers.to_sql("customers", conn, if_exists="replace", index=False)
Location.to_sql("locations", conn, if_exists="replace", index=False)
Orders.to_sql("orders", conn, if_exists="replace", index=False)
Products.to_sql("products", conn, if_exists="replace", index=False)
```

# Tables Exploring

## Customers Table

### Basic quries

#### Select

```python
# selecting All the data from customers Table
query="""
SELECT * 
FROM
customers
"""
df = pd.read_sql_query(query, conn)
df
```

#### limit

```python
# selecting only 10 customers

query="""
select *
from customers
limit 10;
"""
df = pd.read_sql_query(query, conn)
df
```

#### order by

##### asc

```python
# Query write
query='''
select *
from customers
order by [Customer Name]
asc'''
df=pd.read_sql_query(query,conn)
print(df)
```

##### desc

```python
# Query write
query='''
select *
from customers
order by [Customer Name]
desc'''
df=pd.read_sql_query(query,conn)
print(df)
```

# Reports Builing By Using Joins

```python
query="""SELECT c.[Customer Name],
sum(o.Quantity*O.Sales) as Total_Sales
from orders o
join customers c
on o.[Customer ID] = c.[Customer ID]
group by c.[Customer Name]
order by Total_Sales
desc
limit 10;
"""

df = pd.read_sql_query(query, conn)
df
```
