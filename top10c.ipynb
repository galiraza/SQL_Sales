```python
import pandas as pd
import sqlite3
```

```python
conn=sqlite3.connect('mysales.db')

Customers=pd.read_csv('Data/CSV/Customers.csv', encoding='latin-1',sep=";")
Orders=pd.read_csv('Data/CSV/Orders.csv', encoding='latin-1',sep=";")
Products=pd.read_csv('Data/CSV/Products.csv', encoding='latin-1',sep=";")
Location=pd.read_csv('Data/CSV/Location.csv', encoding='latin-1',sep=";")
```

```python
Customers.to_sql("customers",conn,if_exists="replace",index=False)
Orders.to_sql("orders",conn,if_exists="replace",index=False)
Products.to_sql("products",conn,if_exists="replace",index=False)
Location.to_sql("location",conn,if_exists='replace',index=False)
```

```python
query="""SELECT * from customers limit 10;"""
all_customers=pd.read_sql_query(query,conn)
all_customers.describe()
```

```python
query="""select * from orders limit 10"""
all_orders=pd.read_sql_query(query,conn)
```

```python
query="""
SELECT c.[Customer Name],
sum(o.Quantity*o.Sales)
as Total_Sales
from orders o 
join customers c
on c.[Customer ID]=o.[Customer ID]
group by c.[Customer ID]
order by Total_Sales
desc
limit 10;
"""
df=pd.read_sql_query(query,conn)
df
```
